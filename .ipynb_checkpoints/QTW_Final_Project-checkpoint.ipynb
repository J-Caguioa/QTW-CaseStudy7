# Objective

Given an anonymized dataset with a binary target column y, use features x0 through x49 to build a model that would minimize penalty costs on an unseen dataset. 

| | | |
| --- | --- | --- |
| | True Positive: \\$0 | False Negative: <font color='red'>\\$500</font> |
| | False Positive: <font color='#8B0000'>\\$10</font> | True Negative: \\$0 |

Correct labels carry no penalty, while the dollar penalty associated with false negatives outweighs that of false positives by a factor of 50. Thus, our main priority is to create a model that prioritizes avoiding false negatives while also curtailing the number of false positives it produces. Our thought process and analysis of the model performance will be provided along the way where applicable.

# Discussion Points

## Performance Metric

As previously discussed, the driver for this task is minimizing penalty costs incurred by making incorrect classifications. False negatives are much more costly than false positives.

In technical terminology, one good metric for measuring performance in this case would likely be recall (also called sensitivity), which is the fraction of positive cases the model detects. Any positive cases remaining are false negatives — positives that were missed, and the most punitive in this scenario. Thus, tracking which model(s) return the best recall could be especially useful.

However, it's important to note that a model cannot simply call everything positive, as false positives also carry their own monetary cost that could quickly add up. Another useful metric that accounts for this is F1-score, which considers both recall and precision — the fraction of classified positives that are actually positive. Achieving a high F1-score requires reducing both false negatives and false positives, which is also pertinent to this task.

```python
from sklearn.metrics import recall_score, f1_score, confusion_matrix
```

Ultimately however, the monetary cost of misclassifications is the most relevant metric. Tracking recall and F1-score could be helpful from the modeling perspective in the future, giving ideas on what worked best in the event further fine-tuning is needed. But rather than report those metrics, it is more client-relevant to present a metric such as "average penalty cost," where the penalty cost is defined as:

$$PenaltyCost = (FN*500) + (FP*10)$$

The average (here taken as the arithmetic mean) penalty cost of <i>n</i> runs is a quantity that can be compared across models.

```python
def acc_metrics(true, pred):
    print('Recall:', round(recall_score(true, pred), 5))
    print('F1-score:', round(f1_score(true, pred), 5))
    print('-----')
    print('Confusion matrix (top: TN, FP; bottom: FN, TP)')
    print(confusion_matrix(true, pred))
    print('-----')
    print('Penalty Cost: $', penalty_cost(true, pred))
    
def penalty_cost(true, pred):
    tn, fp, fn, tp = confusion_matrix(true, pred).ravel()
    return(fn * 500 + fp * 10)
```

## Splitting the Data

As noted in the EDA, the target variable (as well as the categorical features) contains a class imbalance. Additionally, all features contain missing data. We want to create a model that performs well yet is generalizable to unseen data. One method to achieve this is k-fold cross-validation, where the data is divided into folds that act as both training and validation sets in different iterations. Specifically, to ensure that the class imbalance remains present in each fold, we will employ stratified cross-validation.

However, it would be useful to retain an untouched test set beyond this to estimate the model's performance on "unseen" data, including decisions made in preprocessing such as imputation strategies. This helps prevent data leakage and the possibility of overestimating the chose model's generalizable performance.

The data will be split using a standard ratio as follows:

* 70% training data
* 15% validation data
* 15% test ("unseen") data

For the stratified k-fold cross validation strategy, the 70% training and 15% validation datasets can be thought of as combined because each fold will have a turn at being used as validation data. This 85% combined training-validation set is also used for choosing a preprocessing strategy that is not informed by the holdout 15% test set.

```python
from sklearn.model_selection import train_test_split
```

```python
y = data.y
x = data.drop('y', axis = 1)
```

```python
# Initial split into 85% training-validation and 15% test
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                   stratify = y,
                                                   test_size = 0.15,
                                                   random_state = 42)

# Second split into 70% training and 15% validation
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train,
                                                 stratify = y_train,
                                                 test_size = int(data.shape[0] * 0.15),
                                                 random_state = 42)

print('x_train shape:', x_train.shape)
print('x_val shape', x_val.shape)
print('x_test shape', x_test.shape)
print('y_train shape', y_train.shape)
print('y_val shape', y_val.shape)
print('y_test shape', y_test.shape)
```

| Data | Number of Rows | Percentage |

| Validation | 24000 | 15% | 
| Test | 24000 | 15% |
| <b>Total</b> | <b>160000</b> | <b>100%</b> |

With 136000 of the 160000 total rows, the training and validation data can be combined and further divided into folds. The remaining 24000 rows in the test set will be left untouched until evaluating on a finalized model.

```python
y_data = {'y_train': y_train, 
          'y_test': y_test,
          'y_val': y_val}

fig, ax = plt.subplots(1, len(y_data), figsize = (20, 5))
for i, dataset in enumerate(y_data):
    sns.countplot(y_data[dataset], palette = y_col, ax = ax[i]).set_title(dataset)
plt.tight_layout()
```

Additionally, the graphs above demonstrate successful stratification such that the class imbalance remains intact for the target y in all of the splits.

# Imputation

```python
data.isna().sum()
```

```python
data.isnull().any(axis = 1).sum()
```

As previously established, all columns save for the target have some missing data. A cursory look reveals that no variable has more than 50 missing entries; each individual variable has less than 0.03% of its data missing. Prior to one-hot encoding with dummy variables, 1608 rows in total had at least one missing data entry; this reduces to 1489 afterward. This is approximately 1% of the data. Complicating matters is the anonymized nature of the dataset, as context, variable relations, or background information is often useful to determine whether data for a particular variable is missing completely at random, missing at random, or missing not at random. Each of these categories benefits best from different imputation strategies.

While the simplest solution would be to simply remove any rows with missing data, this should be considered a last resort. Our need to eke out predictive power to help avoid misclassifications, especially false negatives, gives further motivation to investigate whether imputing for these values provides any worthwhile benefit. Furthermore, it would be prudent to have a strategy in place to deal with any data in the unseen test set, or even beyond that, real-life use cases, that contain missing values, as deleting those rows would likely not be a viable option.

(Note that the NA counts are shown for the full dataset to illustrate their prevalence within all features. Decisions on imputation are made on exploring performance of a baseline model built on the training data and evaluated on the validation set; the test set is not considered here.)

## Baseline

An initial attempt is made at seeing the performance on the validation set without any imputation. All rows with any NAs are removed in order to successfully use sklearn. Note that no scaling is necessary for the algorithm used here.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.get_dummies(data)
data_nona = data.dropna()
y_nona = data_nona.y
x_nona = data_nona.drop('y', axis = 1)
x_train_nona, x_test_nona, y_train_nona, y_test_nona = train_test_split(x_nona, y_nona,
                                                   stratify = y_nona,
                                                   test_size = 0.15,
                                                   random_state = 42)

# Second split into 70% training and 15% validation
x_train_nona, x_val_nona, y_train_nona, y_val_nona = train_test_split(x_train_nona, y_train_nona,
                                                 stratify = y_train_nona,
                                                 test_size = int(data_nona.shape[0] * 0.15),
                                                 random_state = 42)
```

```python
print('x_train_nona shape', x_train_nona.shape)
print('x_val_nona shape', x_val_nona.shape)
print('y_train_nona shape', y_train_nona.shape)
print('y_val_nona shape', y_val_nona.shape)
```

```python
clf = RandomForestClassifier(max_depth = 5, n_estimators = 100,
                             random_state = 42)
clf.fit(x_train_nona, y_train_nona)
pred = clf.predict(x_val_nona)
```

```python
acc_metrics(y_val_nona, pred)
```

Our baseline metrics are shown above for a random forest implementation that simply removes all NAs. Given the same algorithm parameters, imputation methods that result in better metrics could be potentially useful.
